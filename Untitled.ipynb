```python
%matplotlib inline
import numpy as np
import csv
import pandas as pd
from scipy.stats import multivariate_normal
from numpy.linalg import inv
from sklearn.metrics.pairwise import rbf_kernel
import matplotlib.pyplot as plt
```

```python
N = 100
n_lines = 5
x = np.linspace(0,10,N).reshape((N,1))
mean = np.zeros(N)
gamma = .05
C = rbf_kernel(x,gamma=gamma)
for i in range(n_lines):
    y = np.random.multivariate_normal(mean,C)
    plt.plot(y)
plt.title("Gamma = {:}".format(gamma))
plt.show()
gamma = .1
C = rbf_kernel(x,gamma=gamma)
for i in range(n_lines):
    y = np.random.multivariate_normal(mean,C)
    plt.plot(y)
plt.title("Gamma = {:}".format(gamma))
plt.show()

gamma = .5
C = rbf_kernel(x,gamma=gamma)
for i in range(n_lines):
    y = np.random.multivariate_normal(mean,C)
    plt.plot(y)
plt.title("Gamma = {:}".format(gamma))
plt.show()
```

# GP fitting

```python
data_x = np.array([-5,-3, -1, 1,3,5]).reshape(-1,1)
data_y = np.array([-.5,1,2,.5,.25,-.45]).reshape(-1,1)
pred = np.linspace(-6,6,100).reshape(-1,1)
gamma = .5
R = rbf_kernel(data_x, pred,gamma = gamma)
C = rbf_kernel(data_x,gamma = gamma)
# print(C)
C_star = rbf_kernel(pred,gamma = gamma)

mu_star = np.dot(np.dot(R.T,inv(C)),data_y)
# print(mu_star.reshape(100))
sigma_star = C_star - np.dot(np.dot(R.T,inv(C)),R)
plt.figure(figsize=(10,10))
for i in range(10):
    y = np.random.multivariate_normal(mu_star.reshape(100),sigma_star)
    plt.plot(pred,y)
plt.plot(data_x,data_y,'ko',markersize=10)
```

```python
#doing the average best line
plt.figure(figsize=(10,10))
n_samp = 100
y = np.zeros((100,n_samp))
for i in range(n_samp):
    y[:,i] = np.random.multivariate_normal(mu_star.reshape(100),sigma_star)
print(np.mean(y,axis=1).shape)
plt.plot(pred,np.mean(y,axis=1))
plt.plot(pred,np.mean(y,axis=1)+np.diag(sigma_star),'k--')
plt.plot(pred,np.mean(y,axis=1)-np.diag(sigma_star),'k--')

plt.plot(data_x,data_y,'ko',markersize=10)
```
